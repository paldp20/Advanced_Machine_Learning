## Task 1

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
data = np.random.randint(0, 21, size=(15, 2))
data
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(data[:,0], data[:,1])
plt.title("Generated data")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
from sklearn.cluster import KMeans

n = 3
kmeans = KMeans(n_clusters=n, n_init=10, random_state=42)
kmeans.fit(data)

labels = kmeans.labels_
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(data[:, 0], data[:, 1], c=labels, label="Data points")
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='x', s=200, c='r', label="Cluster centers")
plt.title(f"K-means Clustering with {n} Clusters")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

```python
cluster_centers = kmeans.cluster_centers_
cluster_centers
```

```python
distances = kmeans.fit_transform(data)
```

```python
distances
```

```python
distances = np.linalg.norm(data - cluster_centers[labels], axis=1)

threshold = np.mean(distances) + 1.5 * np.std(distances)
outliers = data[distances > threshold]

print("Outliers:")
print(outliers)
```

## Task 3

```python
from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples=1000, centers=3, random_state=42)

plt.figure(figsize=(8, 6))
plt.scatter(X[y == 0, 0], X[y == 0, 1], color='red', label='Class 0')
plt.scatter(X[y == 1, 0], X[y == 1, 1], color='blue', label='Class 1')
plt.title("Blobs Dataset")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

```python
from sklearn.cluster import KMeans

n = 3
kmeans = KMeans(n_clusters=n, n_init=10, random_state=42)
kmeans.fit_predict(X)

labels = kmeans.labels_
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X[labels == 0, 0], X[labels == 0, 1], color='red', label='Cluster 0')
plt.scatter(X[labels == 1, 0], X[labels == 1, 1], color='blue', label='Cluster 1')
plt.scatter(X[labels == 2, 0], X[labels == 2, 1], color='green', label='Cluster 2')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], color='black', marker='x', s=200, label='Centroids')
plt.title("K-Means Clustering on Moons Dataset (n_clusters=3)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

```python
cluster_centers = kmeans.cluster_centers_
cluster_centers
```

```python
distances = np.linalg.norm(X - cluster_centers[labels], axis=1)

threshold = np.mean(distances) + 3.5 * np.std(distances)
outliers = X[distances > threshold]

print("Outliers:")
print(outliers)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', marker='o', label='Data Points')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], color='red', marker='x', s=200, label='Centroids')
plt.scatter(outliers[:, 0], outliers[:, 1],  c='b', marker='X', s=200, label="Outlier" )
plt.title("K-Means Clustering with Outliers")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

## Task 4

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler

np.random.seed(0)
data = np.random.randn(300, 2)

scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

dbscan = DBSCAN(eps=0.3, min_samples=5)
labels = dbscan.fit_predict(data_scaled)

plt.figure(figsize=(10, 6))
unique_labels = set(labels)

colors = plt.cm.plasma(np.linspace(0, 1, len(unique_labels)))

for label, color in zip(unique_labels, colors):
    if label == -1:
        plt.scatter(data_scaled[labels == label, 0], data_scaled[labels == label, 1],
                    color='red', marker='x', label='Outliers', s=100)
    else:
        plt.scatter(data_scaled[labels == label, 0], data_scaled[labels == label, 1],
                    color=color, label=f'Cluster {label}', alpha=0.6)

plt.title("DBSCAN Clustering of Random Data Points", fontsize=16)
plt.xlabel("Feature 1 (scaled)", fontsize=14)
plt.ylabel("Feature 2 (scaled)", fontsize=14)

plt.legend()
plt.show()
```

## Task 5

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler

X, _ = make_moons(n_samples=1000, noise=0.1, random_state=42)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

dbscan = DBSCAN(eps=0.2, min_samples=5)
labels = dbscan.fit_predict(X_scaled)

plt.figure(figsize=(8, 6))
unique_labels = set(labels)
colors = plt.cm.Spectral(np.linspace(0, 1, len(unique_labels)))

for k, col in zip(unique_labels, colors):
    if k == -1:
        col = 'red'
    class_member_mask = (labels == k)
    plt.scatter(X_scaled[class_member_mask, 0], X_scaled[class_member_mask, 1],
                color=col, label=f'Cluster {k}' if k != -1 else 'Noise', alpha=0.6, marker='X' if k == -1 else 'o')

plt.title('DBSCAN Clustering on Moons Dataset')
plt.xlabel('Feature 1 (scaled)')
plt.ylabel('Feature 2 (scaled)')
plt.legend()
plt.show()
```
